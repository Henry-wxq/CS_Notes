## 常见图形绘制
- 详见： https://matplotlib.org/stable/gallery/index.html

### 散点图: 分布规律
- plt.scatter()

```python
import matplotlib.pyplot as plt

# 0. 准备数据
x = range(10, 200, 10)
y = range(320, 510, 10)

# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2. 绘制散点图
plt.scatter(x, y)

# 3. 显示图像
plt.show()
```

### 柱状图: 统计，对比
- plt.bar()

```python
# 0. 准备数据
# 电影名字
movie_name = ["The Shawshank Redemption", "The Godfather", "The Dark Knight", "12 Angry Men", "Schindler's List", "The Lord of the Rings", "Pulp Fiction", "Spider-Man", "Pulp Fiction"]
# 横坐标
x = range(len(movie_name))  # 因为电影名字为字符串，需变成数字

# Ratings
rating = [92, 92, 90, 90, 89, 88, 88, 87, 86]

# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2. 绘制柱状图
plt.bar(x, rating, width=0.5, color=['r', 'g', 'b', 'k', 'c', 'm', 'y', 'k', 'g'])

# 2.1 修改x轴的刻度显示
plt.xticks(x, movie_name)

# 2.2 添加网格显示
plt.grid(linestyle="--", alpha=0.5)

# 2.3 添加标题
plt.title('Movie Ratings', fontsize=20)

# 3. 显示图像
plt.show()
```

### 直方图: 统计，分布
- plt.hist()

```python
import numpy as np

rng = np.random.default_rng(19680801)

N_points = 100000
dist1 = rng.standard_normal(N_points)
n_bins = 20

plt.hist(dist1, bins=n_bins)

plt.show()
```

### 饼图: 占比情况
- plt.pie()

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels)

plt.show()
```
